**That's a very broad topic! Exploratory Data Analysis (EDA) Of Real World Data**



1.   **KPI of Real World data:**

>  Total Order,
>  Sold Unite sold,
>  Total Revenue, total discount provided,

2.   **Customer spending**
>  Average spending on a order,
>  Average discount provide on a order



```python

print(f"Total Order Received:        {np.count_nonzero(df.orderid)} orders")
print(f"Total Unite Sold:            {np.sum(df['quantity'])} Unites")
print(f"Total Revenue Generated:     {np.sum(df['salesamount']).round(2)} ")
#print(f"Total Discount Provided:     {np.sum(df['discount']).round(2)} ")
print('Avg time to deliver a order', np.average(df['processingtime']))
print(f"Customer Average Spending On A Order: {(df['salesamount'].sum()/df['orderid'].count()).round(2)} ")

# KPI: percentage of rows where discountapplied == True
yes_percentage = (df['discountapplied'].sum() / len(df)) * 100
print(f"Total Discount approvided : {yes_percentage:.2f}%")
```

```python
# total Revenue by Region
df.groupby('region').agg(Total_revenue = ('salesamount','sum'),Total_order = ('orderid','count')).sort_values('Total_revenue',ascending=False)
```

```python
# The total revenue and order count by category
df.groupby('category').agg(Total_revenue =('salesamount','sum'), Total_order =('orderid','count')).round(2)
```

```python
# Weekly Order received
df.groupby('dayname', as_index= False)['orderid'].count().sort_values('orderid', ascending=False)
```

```python
#Monthly order trend
result = df.groupby('ordermonth', as_index=False).agg(Total_order=('orderid','count'))

month_order =[
     'January', 'February', 'March', 'April', 'May', 'June',
     'July', 'August', 'September', 'October', 'November', 'December']

result['ordermonth']=pd.Categorical(result['ordermonth'], categories=month_order,ordered=True)
result = result.sort_values('ordermonth')
print(result)
```

```python
# Channel
df.groupby('channel').agg(order_count=('orderid','count'),Total_Revenue=('salesamount','sum')).sort_values('order_count',ascending=False)
```

The most significant finding from this analysis is the remarkable consistency in order volume across all seven days of the week. The difference between the peak day (Friday, 307 orders) and the trough day (Monday, 264 orders) is only 43 orders (a variance of approximately 14%).

The most notable trend is the clear dip in order volume during the early summer months, with June (150 orders) and September (151 orders) representing the lowest figures for the entire year. Conversely, the year begins and ends strongly, peaking in January (185 orders) and holding strong in December (178 orders).

* Actionable Strategy: We must initiate a deep-dive into the successful marketing and promotional strategies deployed in January and December. The goal is to adapt these successful Q4/Q1 campaigns to specifically counteract the slump experienced in Q3 (June, July, September), focusing on revitalizing demand during the low-performing summer period.
